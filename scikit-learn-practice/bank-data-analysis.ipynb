```python
numDF = pd.get_dummies(df,columns=['job', 'marital', 'education', 'default', 'housing',
       'loan', 'contact', 'month', 'day_of_week','poutcome',])
```

```python
numDF.dtypes
```

```python
numDF.loc[numDF['y']=='yes','target']=1
numDF.loc[numDF['y']=='no','target']=0
```

```python
numDF.target.value_counts()
```

```python
# 划分训练集与测试集  train_test_split
from sklearn.cross_validation import train_test_split
```

```python
# 70%训练，30%测试
xtrain,xtest,ytrain,ytest = train_test_split(numDF.drop(['y','target'],axis=1),
                                            numDF['target'],train_size =0.7)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
from sklearn import linear_model
```

```python
clf = linear_model.LogisticRegression()
```

```python
clf.fit(xtrain,ytrain)
```

```python
# 开始评估，对每一个模型都有score函数(默认准确率为参数)
clf.score(xtest,ytest)
```

```python
from sklearn.grid_search import GridSearchCV
```

```python
# 比率都是经验，一般不大于1 
param ={"C":[0.01,0.05,0.1,1],"max_iter":[100,200,300]}
```

```python
gsc = GridSearchCV(clf,param,scoring='accuracy',cv=5,verbose=1)
```

```python
gsc.param_grid
```

```python
# 几乎是最快的模型
gsc.fit(xtrain,ytrain)
```

```python
# 多次平均，更加可信  准确率高于0.9几乎很难达到，因为0太多了。准确率是
gsc.grid_scores_
```

```python
gsc.best_params_
```

```python
from sklearn.metrics import precision_recall_curve
```

```python
prob = gsc.predict_proba(xtest)
```

```python
precision,recall,thre =precision_recall_curve(ytest,prob[:,1])
```

```python
precision
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(precision,recall)
```

```python
from sklearn.metrics import classification_report
```

```python
print (classification_report(ytest,gsc.predict(xtest)))
```

```python
# 特征工程师最核心的，所谓特征工程就是造变量
# 特征决定了你的准确率上限，不同模型只是无限逼近上限
# 所有模型接口都是一致的 fit-predict-score-evaluate
# 集成算法是最火的模型，原因是准确率高（如随机森林）
# 涉及到距离计算或梯度下降算法的模型都需要对数据进行归一化或标准化
# （梯度下降，迭代次数，模型收敛【误差达到某个范围，符合想象，符合要求】）

```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# 对数据进行标准化
std = StandardScaler()
```

```python
std.fit(xtrain,ytrain)
```

```python
# 在训练集上标准化并用于测试集，不能在训练集和测试集上各自标准化
xtrain_std = std.transform(xtrain)
xtest_std = std.transform(xtest)
```

```python
clf.fit(xtrain_std,ytrain)
```

```python
clf.score(xtest_std,ytest)
```

```python
print(classification_report(ytest,clf.predict(xtest_std)))
```

```python
print (classification_report(ytest,gsc.predict(xtest)))
```
