```python
numDF = pd.get_dummies(df,columns=['job', 'marital', 'education', 'default', 'housing',
       'loan', 'contact', 'month', 'day_of_week','poutcome',])
```

```python
numDF.dtypes
```

```python
numDF.loc[numDF['y']=='yes','target']=1
numDF.loc[numDF['y']=='no','target']=0
```

```python
numDF.target.value_counts()
```

```python
# åˆ’åˆ†è®­ç»ƒé›†ä¸Žæµ‹è¯•é›†  train_test_split
from sklearn.cross_validation import train_test_split
```

```python
# 70%è®­ç»ƒï¼Œ30%æµ‹è¯•
xtrain,xtest,ytrain,ytest = train_test_split(numDF.drop(['y','target'],axis=1),
                                            numDF['target'],train_size =0.7)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
from sklearn import linear_model
```

```python
clf = linear_model.LogisticRegression()
```

```python
clf.fit(xtrain,ytrain)
```

```python
# å¼€å§‹è¯„ä¼°ï¼Œå¯¹æ¯ä¸€ä¸ªæ¨¡åž‹éƒ½æœ‰scoreå‡½æ•°(é»˜è®¤å‡†ç¡®çŽ‡ä¸ºå‚æ•°)
clf.score(xtest,ytest)
```

```python
from sklearn.grid_search import GridSearchCV
```

```python
# æ¯”çŽ‡éƒ½æ˜¯ç»éªŒï¼Œä¸€èˆ¬ä¸å¤§äºŽ1 
param ={"C":[0.01,0.05,0.1,1],"max_iter":[100,200,300]}
```

```python
gsc = GridSearchCV(clf,param,scoring='accuracy',cv=5,verbose=1)
```

```python
gsc.param_grid
```

```python
# å‡ ä¹Žæ˜¯æœ€å¿«çš„æ¨¡åž‹
gsc.fit(xtrain,ytrain)
```

```python
# å¤šæ¬¡å¹³å‡ï¼Œæ›´åŠ å¯ä¿¡  å‡†ç¡®çŽ‡é«˜äºŽ0.9å‡ ä¹Žå¾ˆéš¾è¾¾åˆ°ï¼Œå› ä¸º0å¤ªå¤šäº†ã€‚å‡†ç¡®çŽ‡æ˜¯
gsc.grid_scores_
```

```python
gsc.best_params_
```

```python
from sklearn.metrics import precision_recall_curve
```

```python
prob = gsc.predict_proba(xtest)
```

```python
precision,recall,thre =precision_recall_curve(ytest,prob[:,1])
```
